```
df['Calculated Discount $'] = df['Order Total'] * (df['Discount %']/100)
df['Discount Match'] = df['Calculated Discount $'].round(2) == df['Discount %'].round()
mismatches = df[df['Discount Match'] == False]
if mismatches.empty:
    print('All discount percentages match the discount dollar amount')
else:
    print(f'Number of mismatches:{mismatches.shape[0]}')
    print(mismatches[['Order No','Order Total','Discount %','Discount $','Calculated Discount $']])
```

Out of all orders, 4510 have mismatches. Discount % is not reliable for calculation; the actual discount is in the Discount $ column, which may include manual or promotional adjustments.
